## Random Horizontal Flipping

### Fashion MNIST Dataset

```python
import torch
import torchvision.transforms as transforms
from torchvision.datasets import FashionMNIST
from matplotlib import pyplot as plt

# Define transformations including Random Horizontal Flip
transform = transforms.Compose([
    transforms.RandomHorizontalFlip(p=0.5),  # 50% chance of flipping
    transforms.ToTensor()  # Convert to tensor for model compatibility
])

# Load the Fashion MNIST dataset with the transform applied
dataset = FashionMNIST(root='./data', train=True, download=True, transform=transform)

# Access a few samples from the dataset
original_dataset = FashionMNIST(root='./data', train=True, download=True, transform=transforms.ToTensor())

# Visualize original and augmented images
fig, axes = plt.subplots(2, 5, figsize=(10, 5))

for i in range(5):
    # Original image
    original_image, label = original_dataset[i]
    axes[0, i].imshow(original_image.squeeze(), cmap='gray')
    axes[0, i].set_title(f"Original\nLabel: {label}")
    axes[0, i].axis('off')

    # Augmented image
    augmented_image, _ = dataset[i]
    axes[1, i].imshow(augmented_image.squeeze(), cmap='gray')
    axes[1, i].set_title("Augmented")
    axes[1, i].axis('off')

plt.tight_layout()
plt.show()
```

## Random Cropping

### Fashion MNIST Dataset

```python
import torchvision.transforms as transforms
from torchvision.datasets import FashionMNIST
from matplotlib import pyplot as plt

# Define transformations with Random Cropping
transform = transforms.Compose([
    transforms.RandomCrop(size=(24, 24)),  # Crop a 24x24 region randomly
    transforms.ToTensor()
])

# Load the Fashion MNIST dataset with the transform applied
dataset = FashionMNIST(root='./data', train=True, download=True, transform=transform)

# Original dataset for comparison
original_dataset = FashionMNIST(root='./data', train=True, download=True, transform=transforms.ToTensor())

# Visualize original and randomly cropped images
fig, axes = plt.subplots(2, 5, figsize=(10, 5))

for i in range(5):
    # Original image
    original_image, label = original_dataset[i]
    axes[0, i].imshow(original_image.squeeze(), cmap='gray')
    axes[0, i].set_title(f"Original\nLabel: {label}")
    axes[0, i].axis('off')

    # Augmented image (random crop)
    cropped_image, _ = dataset[i]
    axes[1, i].imshow(cropped_image.squeeze(), cmap='gray')
    axes[1, i].set_title("Cropped")
    axes[1, i].axis('off')

plt.tight_layout()
plt.show()
```

## RandAugment

### Fashion MNIST Dataset

```python
import torchvision.transforms as transforms
from torchvision.datasets import FashionMNIST
from matplotlib import pyplot as plt

# Define the RandAugment transform
transform = transforms.Compose([
    transforms.RandAugment(num_ops=2, magnitude=9),  # N=2, M=9
    transforms.ToTensor()
])

# Load the Fashion MNIST dataset with RandAugment
dataset = FashionMNIST(root='./data', train=True, download=True, transform=transform)

# Visualize the original and augmented images
original_dataset = FashionMNIST(root='./data', train=True, download=True, transform=transforms.ToTensor())

fig, axes = plt.subplots(2, 5, figsize=(10, 5))

for i in range(5):
    # Original image
    original_image, label = original_dataset[i]
    axes[0, i].imshow(original_image.squeeze(), cmap='gray')
    axes[0, i].set_title(f"Original\nLabel: {label}")
    axes[0, i].axis('off')

    # Augmented image
    augmented_image, _ = dataset[i]
    axes[1, i].imshow(augmented_image.squeeze(), cmap='gray')
    axes[1, i].set_title("Augmented")
    axes[1, i].axis('off')

plt.tight_layout()
plt.show()
```
